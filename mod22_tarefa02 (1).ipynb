#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
df = data.reset_index().tail(60)
````

````python
for ativo in ativos:
  fig = go.Figure(go.Candlestick(
      x=df['Date'],
      open=df[('Open', ativo)],
      high=df[('High', ativo)],
      low=df[('Low', ativo)],
      close=df[('Close', ativo)]
  ))
  fig.update_layout(title_text=ativo)
  fig.show()
````
